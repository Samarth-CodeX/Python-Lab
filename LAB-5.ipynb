```python
graph={
    "A":["C","B"],
    "B":["D","E","F"],
    "C":["F","D"],
    "D":["E","F"],
    "E":["F"],
    "F":[]
}
visited=set()
def dfs(graph,Node,visited):
    if Node not in visited:
        visited.add(Node)
        print(visited)
        for n in graph[Node]:
            dfs(graph,n,visited)

dfs(graph,"A",visited)
```

```python
n = int(input("Enter number of nodes: "))

graph = {}

for i in range(n):
    node = input("Enter node: ")
    neighbors = input("Enter neighbors separated by space: ").split()
    graph[node] = neighbors

root_node = input("Enter The Root Node: ")
goal_node = input("Enter The Goal Node: ")

visited = set()

def dfs(graph, node, goal_node):
    if node in visited:
        return False

    print(node, end=" ")
    visited.add(node)

    if node == goal_node:
        print(f"\nGoal node found - {goal_node}")
        

    for neighbour in graph[node]:
        if dfs(graph, neighbour, goal_node):
            return True

    return False


print("DFS Traversal Order:-")
dfs(graph, root_node, goal_node)
```
